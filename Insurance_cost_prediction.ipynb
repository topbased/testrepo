```python
# Hypothesis Testing
```

```python
## Hypothesis Testing 1:
## H0: Age doesn't determine Premiumprice or H1: Age does determine PremiumPrice.
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['Age'], df['PremiumPrice'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f'Chi-square test statistic: {chi2}')
print(f'p-value: {p}')
```

```python
## ML MODELING
```

```python
## Splitting the features and target
```

```python
X = df.drop(columns = "PremiumPrice",axis = 1)
Y = df["PremiumPrice"]
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
## Splitting the data into Training data and Testing data
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
## Model Training
```

```python
## Linear Regression
```

```python
## Loading the Linear Regression
regressor = LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

```python
## Model Evaluation
```

```python
## Prediction on training data
training_data_prediction = regressor.predict(X_train)
```

```python
## R squared value
```

```python
r2_train = metrics.r2_score(Y_train,training_data_prediction)
print("R squared value :",r2_train)
```

```python
## Prediction on test data
test_data_prediction = regressor.predict(X_test)
```

```python
## R squared value
```

```python
r2_test = metrics.r2_score(Y_test,test_data_prediction)
print("R squared value :",r2_test)
```

```python
## Building a Predictive System

input_data = (45,0,0,0,0,155,57,0,0,0)

## Changing input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

## reshape the array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = regressor.predict(input_data_reshaped)
print(prediction)
```
